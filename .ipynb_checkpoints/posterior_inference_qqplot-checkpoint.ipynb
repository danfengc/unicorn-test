## Ideas behind the posterior inference code check
- posterior samples of MAF for all cases locations given S for all control locations follows a multivarite gaussian ditribution
- the square of the Mahalanobis distance is asymptotically distributed as a chi-square with p degrees of freedom,  where p is the number of variables in the data
- do QQ-plot 

```python
### import the beta for different delta 
import pandas as pd
import numpy as np
```

```python
m_cols = ['m_distance']
mahalanobis_distance = pd.read_csv("/Users/danfengc/projects/unicorn/tests/data/posterior_inference/mahalanobis_distance.txt", 
                                  names = m_cols)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)

sm.qqplot(mahalanobis_distance['m_distance'], 
          stats.chi2, 
          distargs=(5,), 
          line='45',
          ax=ax)

left = 0
top = ax.get_ylim()[1] * 0.75

txt = ax.text(left, 
              top, 
              '', 
              verticalalignment='top')

txt.set_bbox(dict(facecolor='k', 
                  alpha=0.1))
```
