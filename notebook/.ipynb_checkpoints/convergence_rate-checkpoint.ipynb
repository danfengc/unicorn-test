## Check the convergence rate of the MCMC chain



  - given a proposal distrubtion:
    - our goal is to check that for a given proposal distribution, whether the acceptance rate would change with respect to change in 
  in prior distribution, or it would be a constant
  
  - given a prior distribution:
    - difference in acceptance rate with respect to change in proposal distribution
    - mcmc trace plot for different proposal distribution

```python
#  private val delta = 1.0
#  private val ahat = 20.0
#  private val bhat = 30.0
#  private val alpha1 = 4.0
#  private val alpha2 = 5.0
#  private val kappa = 1.0
#  private val scaleSigmaSquare: Double = 1.0
#  private val scalePhi: Double = 1.0
```

### &delta; checker
- changes the value of &delta; while keeping other parameters fixed; the scale parameters for $\sigma^2$ and $\phi$ are denoted as $\theta_{\sigma^2}$ and $\theta_{\phi}$, respectively
    - $\beta$ prior ~ Generalized Logistic distribution(20, 30)
    - $\sigma^2$ prior ~ Gamma(4.0, 1.0)
    - $\phi$ prior ~ Gamma(5.0, 1.0)
    - $\kappa$ = 1
    - $\theta_{\sigma^2}$ = 1.0
    - $\theta_{\phi}$ = 1.0

```python
import pandas as pd
```

```python
col_names = ['delta', 'acceptance_fraction']
accept_frac = pd.read_csv("/Users/danfengc/projects/unicorn/tests/data/mcmc_convergence_test_simple_data/acceptance_frac_delta.txt", 
                         names = col_names, 
                         sep = "\t")
```

- acception fraction for different values of $\delta$; 

```python
accept_frac
```

```python
col_names = ['beta']
beta_phi_100 = pd.read_csv('/Users/danfengc/projects/unicorn/tests/data/mcmc_convergence_test_simple_data/beta_phi_100.0.txt', 
                          names = col_names)

beta_phi_1 = pd.read_csv('/Users/danfengc/projects/unicorn/tests/data/mcmc_convergence_test_simple_data/beta_phi_1.0.txt', 
                          names = col_names)

beta_phi_10Eminus5 = pd.read_csv('/Users/danfengc/projects/unicorn/tests/data/mcmc_convergence_test_simple_data/beta_phi_1.0E-5.txt', 
                          names = col_names)
```

```python
beta_phi_100 = beta_phi_100.reset_index()
beta_phi_1 = beta_phi_1.reset_index()
beta_phi_10Eminus5 = beta_phi_10Eminus5.reset_index()
```

```python
beta_phi_100_trimmed = beta_phi_100[beta_phi_100['index'] % 20 == 0]
beta_phi_1_trimmed = beta_phi_100[beta_phi_1['index'] % 20 == 0]
beta_phi_10Eminus5_trimmed = beta_phi_10Eminus5[beta_phi_10Eminus5['index'] % 20 == 0]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#      plt.title(r'$\alpha > \beta$')

plt.figure(num=None, 
           figsize=(10, 4), 
           dpi=80, 
           facecolor='w', 
           edgecolor='k')

plt.plot(beta_phi_100_trimmed['beta'], 
          color = 'red', 
          alpha = 0.5, 
          label = r'$\delta = 100$')

plt.plot(beta_phi_1_trimmed['beta'], 
          color = 'blue', 
          alpha = 0.5, 
          label = r'$\delta = 1$')

plt.plot(beta_phi_10Eminus5_trimmed['beta'], 
          color = 'green', 
          alpha = 0.5, 
          label = r'$\delta = 10^{-5}$')

plt.ylabel(r'$\beta$')
plt.xlabel("Run")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.title(r'Trace plot for different values of $\delta$ (every 20 samples are shown)')

plt.savefig('/Users/danfengc/projects/unicorn/tests/unicorn-test/plots/trace_beta.pdf', 
            figsize=(12, 4), 
            dpi=80, 
            bbox_inches='tight')
```
